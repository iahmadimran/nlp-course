```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
nlp = spacy.load('en_core_web_sm')
text = 'It is often said that the essence of a thing is not found in the thing itself but in the way that we perceive it through our own experiences.'

doc = nlp(text)

for token in doc:
  if not token.is_stop:
    print(token.text)
```

```python
len(STOP_WORDS)
```

```python
def preprocess(text):
  doc = nlp(text)
  no_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct]
  return " ".join(no_stop_words)
```

```python
preprocess("In accordance with the previously established guidelines, the committee has decided that it would be in the best interest of the organization to postpone the meeting until next Tuesday")
```
