```python
import spacy
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()

words = [
    # Irregular Plurals & Verbs
    "feet", "goose", "geese", "went", "gone", "was", "were",
    
    # Meaning shifts with similar roots
    "organization", "organ", "organic", 
    "policy", "police", "polite",
    
    # Tricky Suffixes (The "i" vs "y" problem)
    "relied", "relying", "reliability",
    "multiplying", "multiplied", "multiplier",
    
    # Ambiguous forms (Is it a noun or a verb?)
    "meeting", "saw", "building",
    
    # Latin/Greek Roots
    "data", "phenomena", "indices", "appendixes"
]

for word in words:
  print(word, '|', stemmer.stem(word))
```

```python
nlp = spacy.load("en_core_web_sm")

sentence = "While building the organization, the police saw that the reliability of the data relied on a polite multiplier multiplying organic phenomena, so they went to a meeting where a goose and geese with many feet were gone, even though the organ policy was multiplied across the indices and appendixes while relying on those who saw the results."

doc = nlp(sentence)

for token in doc:
  print(f'{token} | {token.lemma_}')
  
```
