```python
import spacy
from nltk import sent_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
nlp = spacy.load("en_core_web_sm")

doc1 = nlp("Natural Language Processing is fascinating! or Computers don't speak English. So, we've to learn C, C++, Java, and Python! It is 11 A.M.")

doc2 = nlp("Dr. J.R. Smith visited the U.S.A. at 10 a.m. yesterday. He was accompanied by Prof. Miller from the Dept. of Physics.")

doc3 = nlp("He shouted, 'Watch out!', then ran away. 'Did you see that?', she asked, pointing at the sky.")
```

```python
for sentence in doc1.sents:
  print(sentence)
```

```python
for sentence in doc2.sents:
  print(sentence)
```

```python
for sentence in doc3.sents:
  print(sentence)
```
