```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

text = "Two years ago, he moved down the street to settle down."

doc = nlp(text)

for token in doc:
  print(token, " | ", token.pos_, " | ", spacy.explain(token.pos_), " | ", token.tag_, " | ", spacy.explain(token.pos_))
```

```python
with open('news_story.txt', 'r') as f:
  text = f.read()
```

```python
doc = nlp(text)

nouns = []
nums = []
for token in doc:
  if token.pos_ == 'NOUN':
    nouns.append(token)

  if token.pos_ == 'NUM':
    nums.append(token)
  print(f"{token} | {token.pos_} | {spacy.explain(token.pos_)}")
```

```python
count = doc.count_by(spacy.attrs.POS)
```

```python
for key, value in count.items():
  print(doc.vocab[key].text, "|", value)
```
